# Single Linear Regression


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Advertising1.csv')
df.head()
```

```python
df.corr()
```

Fron above, the information shows two data correlation between together


```python
df.plot(x = 'TV', y = 'Sales', style = 'o')
plt.ylabel('Sales')
```

```python
!pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['TV']]


y = df[['Sales']]
```

```python
lrm = LinearRegression()
lrm.fit(X, y)
```

```python
lrm.intercept_
```

```python
lrm.coef_
```

```python
Predictions = lrm.predict(X)
Predictions
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, Predictions, color = 'blue', linewidth = 3)
plt.show()
```

```python
lrm.predict([[130]])
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y, Predictions)
```

# Multiple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ad2 = pd.read_csv('Advertising2.csv')
df_ad2.head()
```

```python
df_ad2.corr()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_1 = df_ad2[['TV', 'Radio', 'Newspaper', 'Sales']]
y_1 = df_ad2['Sales']
```

```python
lrm = LinearRegression()
lrm.fit(X_1, y_1)
```

```python
Predictions_1 = lrm.predict(X_1)
Predictions_1
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_1, Predictions_1)
```
